```python
import numpy as np
np.__version__
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['figure.dpi'] = 100
np.warnings.filterwarnings('ignore')
```

## Завдання 1

```python
x = np.linspace(-np.pi*8, np.pi*8, num = 101, dtype = np.float32)
```

## Завдання 2

```python
y1 = np.exp(-0.01*x**2)
```

## Завдання 3

```python
y2 = y1 * np.cos(x)
```

## Завдання 4

```python
dy = np.random.uniform(-0.1, 0.1, 101) 
y3 = dy + y2
```

## Завдання 5

```python
print(x.dtype)
print(y1.dtype)
print(y2.dtype)
print(y3.dtype)
y3 = y3.astype(np.float32)
print('-------')
print(y3.dtype)
```

## Завдання 6

```python
y2[40:43] = np.nan
y3[52:55] = np.nan
y2[39:61],y3[39:61]
```

## Завдання 7

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_ylim(-1.15, 1.15)
ax.set_xlim(-np.pi*8 - 2, np.pi*8 + 2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x,y1, 'b--')
ax.plot(x,y2, 'r-')
ax.plot(x,y3, 'g.')
ax.plot(x,-y1, 'b--')
plt.title('Графіки функцій $\pm y_1(x)=e^{-0.01 x^2}$,$ y_2(x)=e^{-0.01 x^2}\cos(x)$\n i "експерементальна" залежність $ y_3(x_i)$')
ax.legend(labels = ('\u00B1$y_1$(x)', '$y_2$(x)', '$y_3$($x_i$)'))
ax.grid(True)
ax.set_xticks([-8*np.pi, -4*np.pi, 0,4*np.pi, 8*np.pi,])
ax.set_xticklabels(['-8$\pi$', '-4$\pi$','0', '4$\pi$', '8$\pi$'])

```
